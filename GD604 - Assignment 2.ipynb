```python
missing_values = df.isna().sum()

print("Missing values per column:")
print(missing_values)
```

```python
df['ADDRESSLINE2'] = df['ADDRESSLINE2'].fillna('Unknown')
df['TERRITORY'] = df['TERRITORY'].fillna('Unknown')
df['POSTALCODE'] = df['POSTALCODE'].fillna('Unknown')
```

```python
df['STATE'] = df['STATE'].fillna(df['STATE'].mode()[0])  
```

```python
missing_values = df.isna().sum()
print("Missing values per column after imputation:")
print(missing_values)
```

```python
import matplotlib.pyplot as plt

# Scatter plot to visualize the relationship between Sales and Profit
plt.figure(figsize=(8, 6))
plt.scatter(df['SALES'], df['PROFIT'], alpha=0.5, color='blue')
plt.title('Sales vs. Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['QUANTITYORDERED'], df['SALES'], alpha=0.5, color='green')
plt.title('Quantity Ordered vs. Sales')
plt.xlabel('Quantity Ordered')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['PRICEEACH'], df['SALES'], alpha=0.5, color='purple')
plt.title('Price Each vs. Sales')
plt.xlabel('Price Each')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
deal_size_distribution = df['DEALSIZE'].value_counts()

deal_size_distribution.plot(kind='pie', autopct='%1.1f%%', colors=['blue', 'green', 'coral'])
plt.title('Distribution of Deal Sizes')
plt.ylabel('')  # Hide the y-label
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='DEALSIZE', y='SALES', data=df, hue='DEALSIZE', palette='pastel', legend=False)
plt.title('Sales Distribution by Deal Size')
plt.xlabel('Deal Size')
plt.ylabel('Sales')
plt.show()
```

```python
from scipy.stats import ttest_ind

small_sales = df[df['DEALSIZE'] == 'Small']['SALES']
large_sales = df[df['DEALSIZE'] == 'Large']['SALES']

t_stat, p_val = ttest_ind(small_sales, large_sales, equal_var=False)

print(f"T-Statistic: {t_stat:.4f}, P-Value: {p_val:.4f}")
```
